```python
#distribution plots
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.distplot(iris['sepal_width'],kde=False,bins=100)
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=iris,kind="kde")

#kind = hex, scatter, reg etc.
```

```python
sns.pairplot(iris,hue='species',palette='coolwarm')
```

```python
sns.rugplot(iris['sepal_length'])
```

```python
sns.barplot(x='sepal_length',y="species",data=iris,estimator="mean")
```

```python
sns.countplot(x='species',data=iris)
```

```python
sns.boxplot(x='sepal_length',y='species',data=iris)
```

```python
sns.heatmap(iris.corr(), cmap="GnBu", linecolor="yellow", linewidth='2',annot=True)
```

```python
sns.heatmap(iris.corr(),cbar=False, xticklabels= False, yticklabels = False)
```

```python
sns.violinplot(x='species',y='sepal_length',data=iris,hue='sepal_length',split=True)
```

```python
sns.stripplot(x='species',y='sepal_width',data=iris,jitter=True) #jitter make its more denser and better
```

```python
sns.violinplot(x='species',y='sepal_length',data=iris)
sns.swarmplot(x='species',y='sepal_length',data=iris)
```

```python
sns.clustermap(iris.corr(),cmap='coolwarm',standard_scale=1)
```

```python
#distribution plots
import seaborn as sns

car_crashes = sns.load_dataset('car_crashes')
car_crashes.head()


```

```python
sns.distplot(car_crashes['alcohol'],kde=False,bins=10)
```

```python
sns.jointplot(x="speeding",y="total",data=car_crashes,kind="reg",color="red")
```

```python
sns.pairplot(car_crashes,hue='total',palette='coolwarm')
```

```python
sns.rugplot(car_crashes['total'])
```

```python
sns.barplot(x='alcohol',y="total",data=car_crashes,estimator="mean")
```

```python
sns.countplot(x='ins_premium',data=car_crashes)
```

```python
sns.boxplot(x='alcohol',y='no_previous',data=car_crashes)
```

```python
sns.heatmap(car_crashes.corr(), cmap="GnBu", linecolor="blue", linewidth='2',annot=True)
```

```python


sns.violinplot(x='speeding',y='not_distracted',data=car_crashes)
```

```python

sns.stripplot(x='speeding',y='alcohol',data=car_crashes,jitter=True)
```
